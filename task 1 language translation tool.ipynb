```python
# Step 1: Install required library (only needs to run once)
# If running in Colab/Jupyter, uncomment the line below and run it first:
# !pip install deep-translator gTTS -q

from deep_translator import GoogleTranslator
from gtts import gTTS

# Languages supported (name -> code)
LANGUAGES = {
    "1": ("English", "en"),
    "2": ("Urdu", "ur"),
    "3": ("Arabic", "ar"),
    "4": ("French", "fr"),
    "5": ("Spanish", "es"),
    "6": ("German", "de"),
    "7": ("Chinese (Simplified)", "zh-CN"),
    "8": ("Hindi", "hi"),
    "9": ("Turkish", "tr"),
    "10": ("Russian", "ru"),
}


def choose_language(prompt):
    print(prompt)
    for key, (name, _) in LANGUAGES.items():
        print(f"  {key}. {name}")
    choice = input("Enter number: ").strip()
    while choice not in LANGUAGES:
        choice = input("Invalid choice, enter a number from the list above: ").strip()
    return LANGUAGES[choice][1]


def main():
    print("=" * 50)
    print("CodeAlpha AI Internship - Task 1")
    print("    Language Translation Tool")
    print("=" * 50)

    # Step 1: Get text from the user
    text = input("\nEnter the text you want to translate:\n> ")

    # Step 2: Choose source and target language
    source_lang = choose_language("\nSelect SOURCE language (or type 'auto' for auto-detect):")
    target_lang = choose_language("\nSelect TARGET language:")

    # Step 3: Send text to the translation API and get result
    try:
        translated_text = GoogleTranslator(source=source_lang, target=target_lang).translate(text)
    except Exception as e:
        print(f"\n Translation failed: {e}")
        return

    # Step 4: Display the translated text
    print("\n" + "=" * 50)
    print(" Translated Text:")
    print(translated_text)
    print("=" * 50)

    # Step 5 (Bonus): Text-to-Speech - listen to the translation
    listen = input("\nWould you like to hear the translation out loud? (y/n): ").strip().lower()
    if listen == "y":
        try:
            tts = gTTS(text=translated_text, lang=target_lang.split("-")[0])
            tts.save("translated_audio.mp3")
            print("🔊 Audio saved as 'translated_audio.mp3' - play this file to listen.")

            # If running inside Jupyter/Colab, also try to play it directly
            try:
                from IPython.display import Audio, display
                display(Audio("translated_audio.mp3", autoplay=True))
            except ImportError:
                pass  # not running in a notebook, that's fine

        except Exception as e:
            print(f"🔈 Text-to-speech not available for this language: {e}")


if __name__ == "__main__":
    main()
```
